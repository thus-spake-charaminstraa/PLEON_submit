```python
DATA_PATH = "/content/drive/Othercomputers/MacBookPro/plantDetection/keywords.txt"

import os
import csv
import pandas as pd
import numpy as np
```

plant dictionary
key : 학명 value : class

```python
txt = open(DATA_PATH, "r+")

plantDict = {}
plantIndex = 0

while True :
    line = txt.readline()
    if not line:
        break
    #plantDict[line.strip()] = 80 + plantIndex
    plantDict[line.strip()] = plantIndex
    plantIndex += 1

txt.close()

print(plantDict)
```

Image파일의 path csv 만들기

```python
head = ['image_path','class']
IMAGE_PATH = '/content/drive/Othercomputers/MacBookPro/plantDetection/datasets/plant/images/train2022'
img_list = os.listdir(IMAGE_PATH)

#data.csv 만들기
with open('/content/drive/Othercomputers/MacBookPro/plantDetection/data.csv', 'w', newline='') as f:
    w = csv.writer(f)
    w.writerow(head)
    for p in plantDict.keys():
      for img in img_list :
        if p == img.split('_')[0] : 
          temp = [ (IMAGE_PATH + '/' + img), plantDict[p]]
          w.writerow(temp)
```

TXT파일 라벨 수정


```python
head = ['txt_path','class']
TXT_PATH = '/content/drive/Othercomputers/MacBookPro/plantDetection/datasets/plant/labels/train0706'

#ex = '/content/drive/Othercomputers/MacBookPro/plantDetection/datasets/plant/labels/train2022/Zygocactus truncatus_4110.txt'

txt_list = os.listdir(TXT_PATH)

print(len(txt_list))

check = 0
#txt.csv 만들기
with open('/content/drive/Othercomputers/MacBookPro/plantDetection/txt_path.csv', 'w', newline='') as f:
    w = csv.writer(f)
    w.writerow(head)
    for p in plantDict.keys():
      for txt in txt_list :
        if p == txt.split('_')[0] : 
          temp = [ (TXT_PATH + '/' + txt), plantDict[p]]
          w.writerow(temp)
          check += 1
print(check)
```

```python
#df = pd.read_csv('/content/drive/Othercomputers/MacBookPro/plantDetection/txt_path.csv')
#path_list = df['txt_path']

# ex = '/content/drive/Othercomputers/MacBookPro/plantDetection/datasets/plant/labels/train2022/Zygocactus truncatus_4110.txt'
# temp = ex.split('/')[-1]
# temp = temp.split('_')[0]
# print(temp)

#print(len(path_list))
#print(check)
TXT_PATH = '/content/drive/Othercomputers/MacBookPro/plantDetection/datasets/plant/labels/train0706'

txt_list = os.listdir(TXT_PATH)

#print(len(txt_list))

txt = open('/content/drive/Othercomputers/MacBookPro/plantDetection/datasets/plant/labels/train0706/' + txt_list[0], "r+")
lines = txt.readlines()
txt.close()

#print(txt_list)
#print(lines)

sum = 0
for i in range(len(txt_list)) :
  txt_name = txt_list[i]
  txt = open('/content/drive/Othercomputers/MacBookPro/plantDetection/datasets/plant/labels/train0706/' + txt_list[i], "r+")
  #print(txt_list[i])
  lines = txt.readlines()
  txt.close()
  sum += 1
  #print(lines)

  pName =  txt_name.split('/')[-1]
  pName = pName.split('_')[0]
  pre_class = 58
  if pName == 'Hoya carnosa ':
    pName = 'Hoya carnosa'
  if pName == 'Anthurium andraeanum Linden ex André':
    pName = 'Anthurium andraeanum Linden ex André'
  post_class = plantDict[pName]
  #print(post_class)

  fw = open('/content/drive/Othercomputers/MacBookPro/plantDetection/datasets/plant/labels/train0706/' + txt_list[i], 'w')
  for line in lines:
    fw.write(line.replace(str(pre_class), str(post_class)))
  fw.close()
   
print(sum)
```
